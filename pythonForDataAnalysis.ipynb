```python
# Visualization with Pandas
import pandas as pd
df_population_raw = pd.read_csv('population_total.csv')
df_population_raw
```

```python
# Dropping the null values
df_population_raw.dropna(inplace=True)
```

```python
df_pivot = df_population_raw.pivot(index='year', columns='country',
                        values='population')
```

```python
df_pivot = df_pivot[['United States', 'India', 'China', 'Indonesia','Brazil']]
```

```python
df_pivot
```

```python
# Visualization using pandas
# using the line chart for the visualization
df_pivot.plot(kind='line', xlabel='Year',
              ylabel='Population', title='Population(1955-2020)', figsize=(8,4))
```

```python
# Bar Plot
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
```

```python
# To transpose
df_pivot_2020 = df_pivot_2020.T
```

```python
 df_pivot_2020.plot(kind='bar', color='orange',
                    xlabel='Year', ylabel='Population',
                    title= 'Population (2020)')
```

```python
df_pivot_sample = df_pivot[df_pivot.index.isin([1980,2000, 2010, 2020])]
df_pivot_sample
```

```python
df_pivot_sample.plot(kind='bar', xlabel='Year',
                     ylabel='Population')
```

```python
# Pie Chart
df_pivot_2020 = df_pivot_2020.T
df_pivot_2020
```

```python
df_pivot_2020.rename(columns={2020:'2020'},inplace=True)
```

```python
df_pivot_2020.plot(kind='pie',y='2020',
                   title='Population(%)')
```
